```python
import numpy as np
import pandas as pd
from astropy.table import Table

from matplotlib import pyplot as plt
```

```python
fn_mock = '../data/OneEightSky_z0.45_box2b_hr/wmap.027.galaxies.dat'
```

```python
columns_to_load = ['isub', 'l', 'b', 'z_obs', 'Mstar', 'sfr']
all_columns = ['isub', 'l', 'b', 'rr', 'vmax', 'z_true', 'z_obs', 'Mstar', 
                  'sfr', 'u', 'V', 'g', 'r', 'i', 'z', 'Y', 'J', 'H', 'K', 
                  'L', 'M', 'Age', 'Z']
    
# Find indices of columns to load
usecols = [all_columns.index(col) for col in columns_to_load if col in all_columns]

# Load only the selected columns
mock_mag = pd.read_csv(fn_mock, 
              sep='\s+',      # Whitespace delimiter
              header=0,    # No header in file
              usecols=usecols, # Only read specified columns
              names=columns_to_load)  # Assign column names
```

```python
mock_mag['z_obs'].values
```

```python
plt.hist(mock_mag['z_obs'], bins=30)
plt.xlabel('z')
```

```python
plt.hist(np.log10(mock_mag['Mstar']), bins=30)
plt.xlabel('log(Mstar)')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(mock_mag['l'], mock_mag['b'], s=0.1, alpha=0.01, color='k')
plt.xlabel('l')
plt.ylabel('b')
```
